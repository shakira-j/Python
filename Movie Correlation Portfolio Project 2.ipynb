```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams ['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

#reading in the data

df = pd.read_csv('/Users/shakirajohnson/Downloads/movies.csv')
```

```python
#Checking to see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
df=df.dropna()
```

```python
#confirming missing data has been dropped
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
#Sorting in descending order by gross
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#checking the data types of the columns
df.dtypes
```

```python
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#creating scatter plot with buget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')


plt.show()
```

```python
df.head()
```

```python
# plotting the budget vs gross using seaborn to see if budget and gross are correlated 

sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
df.corr(method='pearson') #high correlation between budget and gross
```

```python
#using heatmap to visualize correlation for easier identification
correlation_matrix = df.corr(method='pearson')

plt.title('Correlation Matrix for Movie Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

sns.heatmap(correlation_matrix, annot=True)

plt.show
```

```python
df.head()
```

```python
#creating unique identifiers in order to find company correlation to gross revenue

df_categorized = df

for col_name in df_categorized.columns: 
    if(df_categorized[col_name].dtype == 'object'):
        df_categorized[col_name] = df_categorized[col_name].astype('category')
        df_categorized[col_name] = df_categorized[col_name].cat.codes
        
df_categorized.head()
```

```python
correlation_matrix = df_categorized.corr(method='pearson')

plt.title('Correlation Matrix for Movie Features')

sns.heatmap(correlation_matrix, annot=True)

plt.show
```

```python
df_categorized.corr()
```

```python
#easily see correlation 
correlation_mat = df_categorized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
